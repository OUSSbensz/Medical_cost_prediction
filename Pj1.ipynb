# DATA LOADING

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("insurance.csv")
df
```

# Data Preprocessing

```python
df.dtypes # To know the type of our variables
```

```python
df.isnull().sum() # to detect the missing values
```

```python
#Grouping Numerical feature
numerical=df.select_dtypes(include=['int','float']).columns
numerical_data=df[numerical]
numerical_data
```

```python
#Grouping Categorical features
Categorical=df.select_dtypes(include='object').columns
Categorical_data=df[Categorical]
Categorical_data
```

```python
#One hot Encoding
data_encoded = pd.get_dummies(Categorical_data)
data_encoded
```

```python
# Convert boolean values to integers (True: 1, False: 0)
df_boolean = data_encoded.select_dtypes(include=bool).astype(int)
df_boolean
```

```python
# Combine datasets horizontally based on the 'ID' column
combined_df = pd.concat([numerical_data, df_boolean], axis=1)
combined_df
```

```python
sns.regplot(x="bmi ",y="charges",data=combined_df)
plt.show
```
